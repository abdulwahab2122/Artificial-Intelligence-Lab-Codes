```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('fifa_dataset.csv')
df.head()

# Correlation between all the columns with each other
df.corr()

# Heatmap using seaborn library
sns.heatmap(df.corr(), annot=False)

# Pairs Plots with Seaborn for Age, Overall and Potential
sns.pairplot(df[['Age', 'Overall', 'Potential']])
plt.show()


```
